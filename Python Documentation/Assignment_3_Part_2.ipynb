### Introduction
The Titanic dataset contains information on passengers who were onboard the Titanic when it sank on April 15, 1912. The dataset contains 891 rows and 12 columns, and it has information about each passenger's survival status, age, gender, passenger class, fare, and other attributes.

In this exploratory data analysis (EDA), we will answer three hypotheses:

**Hypothesis 1 :** Survival rate is associated with the class of passenger.

**Hypothesis 2 :** Survival rate is associated with the gender of the passenger.

**Hypothesis 3 :** Survival rate is associated with the age of the passenger.

Let's get started by importing necessary libraries and loading the dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

titanic = pd.read_csv("titanic.csv")  # Loading the dataset into pandas dataframe

```

Now, let's have a look at the first few rows of the dataset to get a sense of the data.

```python
titanic.head()  #displaying the first five data values from dataframe.
```

### Hypothesis 1: Survival rate is associated with the class of passenger.

To test this hypothesis, we will create a bar chart that shows the survival rate of passengers in each class.

```python
# Creating a bar chart showing the survival rate by passenger class using seaborn barplot code.
sns.barplot(x="Pclass", y="Survived", data=titanic)

# Adding respective labels to the chart
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()
```

The resulting chart shows that passengers in first class had the highest survival rate, followed by second class, and third class. This suggests that there is an association between passenger class and survival rate.

### Hypothesis 2: Survival rate is associated with the gender of the passenger.

To test this hypothesis, we will create a bar chart that shows the survival rate of male and female passengers.

```python
# Create a bar chart showing the survival rate by gender
sns.barplot(x="Sex", y="Survived", data=titanic)

# Add labels to the chart
plt.title("Survival Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.show()
```

The resulting chart shows that female passengers had a much higher survival rate than male passengers. This suggests that there is an association between gender and survival rate.

### Hypothesis 3: Survival rate is associated with the age of the passenger.

To test this hypothesis, we will create a scatter plot that shows the relationship between age and survival.

```python
# Creating a scatter plot showing the relationship between age and survival
sns.scatterplot(x="Age", y="Survived", data=titanic)

# Adding labels to the chart
plt.title("Survival Rate by Age")
plt.xlabel("Age")
plt.ylabel("Survival Rate")
plt.show()
```

The resulting chart shows that there does not appear to be a clear association between age and survival rate. However, we can see that there are some spikes in the survival rate for certain age ranges, particularly children and young adults.
